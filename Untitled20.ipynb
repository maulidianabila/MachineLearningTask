```python
%%writefile app.py
import streamlit as st
import torch
import numpy as np
import cv2
from PIL import Image

st.title("🦁 Deteksi Hewan Menggunakan YOLOv5")

uploaded_file = st.file_uploader("Upload gambar hewan", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Gambar yang diunggah", use_column_width=True)

    # Load model YOLOv5
    model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # Model pre-trained

    img = np.array(image)

    # Deteksi
    results = model(img)
    results.render()  # Tambahkan bounding box ke image

    # Tampilkan hasil
    st.image(results.ims[0], caption='Hasil Deteksi Hewan', use_column_width=True)

    # Tampilkan label deteksi
    st.write("Objek terdeteksi:")
    st.table(results.pandas().xyxy[0][['name', 'confidence']])
```
